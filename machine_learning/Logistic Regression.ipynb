## Logistic Regression

```python
from numpy import *
from pylab import *
import numpy.random as rnd
```

#### Prepare for data

```python
datas_around_0_0 = array([rnd.normal([0, 0], 3) for x in range(50)])
datas_around_10_10 = array([rnd.normal([10, 10], 3) for x in range(50)])
datas_around_5_5 = array([rnd.normal([5, 5], 2) for x in range(100)])
datas_c1 = concatenate([datas_around_0_0, datas_around_10_10])
datas_c2 = datas_around_5_5
```

#### Separate to training and test set

```python
shuffle(datas_c1)
shuffle(datas_c2)

train_c1, train_c2 = datas_c1[:60], datas_c2[:60]
test_c1, test_c2 = datas_c1[60:], datas_c2[60:]
```

#### Plot training data

```python
scatter(train_c1[:, 0], train_c1[:, 1], color="red", label="Class 1")
scatter(train_c2[:, 0], train_c2[:, 1], color="blue", label="Class 2")
legend()
```

#### Feature function of logistic regression using gaussian basis:
$$ f(d_i) = f(x_i, y_i) = \sigma(w_1 \mbox{GaussianBasis}(x_1, y_1) + w_2  \mbox{GaussianBasis}(x_2, y_2) + \mbox{...} + w_n \mbox{GaussianBasis}(x_n, y_n)) $$

#### Where \sigma is sigmod function:
$$ \sigma(x) = \frac{1}{1 + e^{-x}} $$ 

#### Change this into vectorized expression:
$$ \phi_i = \phi(d_i) = \phi(x_i, y_i) = (x_i, y_i, x_i y_i, x_i^2, y_i^2)$$
$$ {\mathbf w} = (w_1, w_2, ..., w_5) $$
$$ f(d_i) = \sigma({\mathbf w}^T {\mathbf \phi}) $$

#### Likelihood of feature function:
$$ L({\mathbf w}|{\mathbf D}, {\mathbf t}) = \prod_i f(d_i)^t (1-f(d_i))^{1-t} $$

#### Log Likelihood
$$ ln L({\mathbf w}|{\mathbf D}, {\mathbf t}) = \sum_i t \ln f(d_i) + (1-t) \ln (1-f(d_i)) $$

#### Differentiate
\begin{aligned}
        \frac{\mathrm{d}}{\mathrm{d}x}\sigma(x) &= \frac{\mathrm{d}}{\mathrm{d}x}\frac{1}{1+\exp(-x)} \\
        &= \frac{\exp(-x)}{(1+\exp(-x))^2} \\
        &= \frac{1}{1+\exp(-x)}\left\{1-\frac{1}{1+\exp(-x)}\right\} \\
        &= \sigma(x)(1-\sigma(x))
        \end{aligned}

#### Differentiate the log likelihood
\begin{aligned}
        \frac{\mathrm{d}}{\mathrm{d}\mathbf{w}}\ln \sigma(\mathbf{w}^T\boldsymbol{\phi}) 
        &= \frac{1}{\sigma(\mathbf{w}^T\boldsymbol{\phi})}\frac{\mathrm{d}}{\mathrm{d}\mathbf{w}}\sigma(\mathbf{w}^T\boldsymbol{\phi}) \\
        &= \frac{1}{\sigma(\mathbf{w}^T\boldsymbol{\phi})}\sigma(\mathbf{w}^T\boldsymbol{\phi})\left\{1-\sigma(\mathbf{w}^T\boldsymbol{\phi})\right\}\frac{\mathrm{d}}{\mathrm{d}\mathbf{w}}\mathbf{w}^T\boldsymbol{\phi} \\
        &= \frac{1}{\sigma(\mathbf{w}^T\boldsymbol{\phi})}\sigma(\mathbf{w}^T\boldsymbol{\phi})\left\{1-\sigma(\mathbf{w}^T\boldsymbol{\phi})\right\} \boldsymbol{\phi} \\
        &= \{1-\sigma(\mathbf{w}^T\boldsymbol{\phi})\}\boldsymbol{\phi}
        \end{aligned}

#### In the same way
$$ \frac{\mathrm{d}}{\mathrm{d}\mathbf{w}}\ln \{1-\sigma(\mathbf{w}^T\boldsymbol{\phi})\} = -\sigma(\mathbf{w}^T\boldsymbol{\phi})\boldsymbol{\phi} $$


#### So, the differentiation of log likelihood is
\begin{aligned}
        &\frac{\mathrm{d}}{\mathrm{d}\mathbf{w}}\ln L(\mathbf{w}|D, \mathbf{t}) \\
        =&\sum_i\left[t_i(1-\sigma(\mathbf{w}^T\boldsymbol{\phi}_i))\boldsymbol{\phi}_i+(1-t_i)(0-\sigma(\mathbf{w}^T\boldsymbol{\phi}_i))\boldsymbol{\phi}_i\right] \\
    =& \sum_i (t_i-\sigma(\mathbf{w}^T\boldsymbol{\phi}_i))\boldsymbol{\phi}_i \\
        \end{aligned}

#### Let 
\begin{aligned}
        \mathbf{t}&=(t_1,t_2,\ldots,t_n)^T\\
        \mathbf{y}&=(\sigma(\mathbf{w}^T\boldsymbol{\phi}_1),\sigma(\mathbf{w}^T\boldsymbol{\phi}_2),\ldots,\sigma(\mathbf{w}^T\boldsymbol{\phi}_n))^T\\
        \mathbf{X}&=(\boldsymbol{\phi}_1,\boldsymbol{\phi}_2,\ldots,\boldsymbol{\phi}_n)^T
        \end{aligned}

#### Then
$$ \nabla \ln L(\mathbf{w}|D) = \frac{\mathrm{d}}{\mathrm{d}\mathbf{w}}\ln L(\mathbf{w}|D) = \mathbf{X}^T(\mathbf{t}-\mathbf{y}) $$

#### Differntiate $\nabla \ln L(\mathbf{w}|D)$ agian
\begin{aligned}
    \frac{\mathrm{d}^2}{\mathrm{d}^2\mathbf{w}} \ln L(\mathbf{w}|D) &= \frac{\mathrm{d}}{\mathrm{d}\mathbf{w}} \sum_i (t_i-\sigma(\mathbf{w}^T\boldsymbol{\phi}_i))\boldsymbol{\phi}_i \\
    &= - \frac{\mathrm{d}}{\mathrm{d}\mathbf{w}} \sum_i \sigma(\mathbf{w}^T\boldsymbol{\phi}_i)\boldsymbol{\phi}_i \\
    &= - \sum_i \sigma(\mathbf{w}^T\boldsymbol{\phi}_i)(1 - \sigma(\mathbf{w}^T\boldsymbol{\phi}_i)) ( \frac{\mathrm{d}}{\mathrm{d}\mathbf{w}} \mathbf{w}^T\boldsymbol{\phi}_i ) \boldsymbol{\phi}_i^T \\
    &= - \sum_i \sigma(\mathbf{w}^T\boldsymbol{\phi}_i)(1 - \sigma(\mathbf{w}^T\boldsymbol{\phi}_i)) \boldsymbol{\phi}_i  \boldsymbol{\phi}_i^T
\end{aligned}

#### Which is actually a hessian matrix

\begin{aligned}
        \mathbf{H}&=-\mathbf{X}^T\mathbf{R}\mathbf{X}\\
        \mathbf{R}&=\begin{pmatrix}
        \sigma(\mathbf{w}^T\boldsymbol{\phi}_1)(1-\sigma(\mathbf{w}^T\boldsymbol{\phi}_1)) & \cdots & 0 \\
        \vdots & \ddots & \vdots \\
        0 & \cdots & \sigma(\mathbf{w}^T\boldsymbol{\phi}_n)(1-\sigma(\mathbf{w}^T\boldsymbol{\phi}_n)) 
        \end{pmatrix}
        \end{aligned}

## Parameter Estimation

#### Generally we have a lot of methods for estimating parameters
- Gradient Descent
- Stochastic gradient descent (SGD)
- Newton-Raphson method 
- Broyden–Fletcher–Goldfarb–Shanno algorithm (BFGS)
- BHHH
- AdaGrad
- ...

### Newton-Raphson method 

$$ \mathbf{w}_{k+1} = \mathbf{w}_k - \mathbf{H}_k^{-1}\nabla \ln L_k $$

#### Also, intutively, for $f'(x)=0$
$$ x_{k+1} = x_k - \frac{f'(x_k)}{f''(x_k)} $$

#### In this case
\begin{aligned}
 \mathbf{w}_{k+1} =& \mathbf{w}_k - \frac{\mathbf{X}^T(\mathbf{t}-\mathbf{y}_k)}{\frac{\mathrm{d}^2}{\mathrm{d}^2\mathbf{w}} \ln L(\mathbf{w}|D)} \\
 =& \mathbf{w}_k + (\mathbf{X}^T\mathbf{R}_k\mathbf{X})^{-1}\mathbf{X}^T(\mathbf{t}-\mathbf{y}_k)
\end{aligned}

#### Where 
$$ R_{ii} = \sigma(\mathbf{w}^T\boldsymbol{\phi}_i)(1-\sigma(\mathbf{w}^T\boldsymbol{\phi}_i)) $$

### Implementation of feature function

```python
import math
def sigmod(x):
    return 1.0 / (1 + math.exp(-x))
```

```python
BASIS_SIGMA = 3.0
BASIS_COUNT = 4
DIM = BASIS_COUNT*BASIS_COUNT

basis_center = [(x, y) for x in linspace(-10, 20, BASIS_COUNT)
                       for y in linspace(-10, 20, BASIS_COUNT)]


def gaussian_basis(x, y, cx, cy):
    return math.exp(-((x-cx)**2 + (y-cy)**2)/(2*BASIS_SIGMA**2))

def phi_i(d_i):
    x, y = d_i
    return array([gaussian_basis(x, y, cx, cy) for (cx,cy) in basis_center])

def f(w, phi):
    return sigmod(w.T.dot(phi))
```

#### $ \nabla \ln L(\mathbf{w}|D)$

### Implementation Newton-Raphson method

```python
def make_X(D):
    return array([phi_i(d) for d in D])

def make_y(w, X):
    return array([f(w, phi) for phi in X])

def make_gradient(w, X, t, y):
    return X.T.dot(t-y)

def make_H(w, X, y):
    R_iis = array([y_i*(1-y_i) for y_i in y])
    R = diag(R_iis)
    return - X.T.dot(R).dot(X)
```

```python
def update(w, X, t):
    y = make_y(w, X)
    gradient = make_gradient(w, X, t, y)
    H = make_H(w, X, y)
    return w - linalg.solve(H, gradient)
```

```python
D = concatenate([train_c1, train_c2])
t = array( [1]*len(train_c1) + [0]*len(train_c2) )
X = make_X(D)
```

```python
init_w = array([1]*DIM)
```

#### Update w 5 times

```python
w = init_w
for i in range(5):
    w = update(w, X, t)
    xspace, yspace = meshgrid(linspace(-10, 20, 100), linspace(-10, 20, 100))
    zs = vectorize(lambda x,y: f(w, phi_i((x,y))))(xspace, yspace)
    pcolor(xspace, yspace, zs, alpha=0.3)
    xlim(-10, 20)
    ylim(-10, 20)
    scatter(train_c1[:, 0], train_c1[:, 1], color="red", label="Class 1")
    scatter(train_c2[:, 0], train_c2[:, 1], color="blue", label="Class 2")
    show()
```

### OverflowError !!!

- #### For exp(x), when |x| >= 710, overflows or underflows,
- #### For sigmod(x), when |x| >= 36, loss of trailing digits 

#### To avoid overflow and underflow, use following function $$ \sigma(x) = \frac{1}{2} (1 + \tanh \frac{x}{2} ) $$

```python
import math
def sigmod(x):
    return (1 + math.tanh(x/2)) / 2
```

#### Plot this function

```python
xs = linspace(-10, 10)
plot(xs, vectorize(sigmod)(xs))
title("(1 + tanh(x/2)) / 2", color="gray")
show()
```

#### Seems good even only 5 iterations

#### Let's see the accuracy

```python
def determine(sigmod_value):
    return 1 if sigmod_value >= 0.5 else 0
```

```python
predictions = [determine(f(w, phi_i(d))) for d in concatenate([test_c1, test_c2])]
correct_labels = [1] * len(test_c1) + [0] * len(test_c2)
```

```python
hits = len(filter(lambda d: d[0] == d[1], zip(predictions, correct_labels)))
print "Accuracy: %f ( %d / %d )" % (float(hits)/len(correct_labels), hits, len(correct_labels))
```
