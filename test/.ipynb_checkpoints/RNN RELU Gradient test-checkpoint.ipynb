# RNN RELU Gradient test

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn
```

```python
import theano
import theano.tensor as T
import numpy as np
from collections import OrderedDict
```

```python
import sys
sys.path.append("/home/hadoop/deepy")
from deepy.util import smart_replace_graph as SRG
```

```python
def relu(x):
    return x * (x > 0)
```

### One layer

```python
x = T.vector("x")
b = T.vector("b")
w = T.matrix("w")

z1 = T.dot(x, w) + b
s1 = relu(z1)
g1 = T.grad(T.sum(s1), x)
g1_simple = SRG(g1, OrderedDict({T.sum(s1): T.scalar("sum_s1"), s1: T.vector("s1"), z1: T.vector("z1")}))
print theano.pp(g1_simple)
gw1 = T.grad(T.sum(s1), w)
gw1_simple = SRG(gw1, OrderedDict({T.sum(s1): T.scalar("sum_s1"), s1: T.vector("s1"), z1: T.vector("z1")}))
print theano.pp(gw1_simple)
```

```python
grad_f = theano.function([x, w, b], g1)
x_var = np.array([2., 1., 3.])
w_var = np.array([[-1,2],[1,4],[-2, 5]])
b_var = np.zeros(2)
print "theano grad:", grad_f(x_var, w_var, b_var)
print "hand-crafted grad:", ((x_var.dot(w_var) + b_var) > 0).dot(w_var.T)
```

### $$ \nabla x = w(w^Tx + b > 0)  $$

```python
gw_f = theano.function([x, w, b], gw1)
x_var = np.array([2., 1., 3.])
w_var = np.array([[-1,2],[1,4],[-2, 5]])
b_var = np.zeros(2)
print "theano grad:", gw_f(x_var, w_var, b_var)
print "hand-crafted grad:", np.vstack([x_var,x_var]).T * ((x_var.dot(w_var) + b_var) > 0)
```

### Two layer

```python
x = T.vector("x")
b = T.vector("b")
w = T.matrix("w")

def recurent(s):
    return relu(T.dot(s, w) + b)

z1 = T.dot(x, w) + b
s1 = relu(z1)

z2 = T.dot(s1, w) + b
s2 = relu(z2)

g2 = T.grad(T.sum(s2), x)
g2 = SRG(g2, OrderedDict({T.sum(s2): T.scalar("sum_s2"), s2: T.vector("s2"), z2: T.vector("z2"),
                          s1: T.vector("s1"), z1: T.vector("z1")}))
gw2 = T.grad(T.sum(s2), w)
gw2_simple = SRG(gw2, OrderedDict({T.sum(s2): T.scalar("sum_s2"), s2: T.vector("s2"), z2: T.vector("z2"),
                          s1: T.vector("s1"), z1: T.vector("z1")}))
print theano.pp(g2)
print theano.pp(gw2_simple)
```

#### gradient of w:

(DimShuffle{0, x}(x) \dot (z2 > 0) \dot w.T) * gt(z1, TensorConstant{0}) ) + (DimShuffle{0, x}(s1) \dot (z2 > 0))

### $$ \nabla x  = w( ( \ w(z_2 > 0 )\  )(z_1 > 0) )  $$

### Expansion

 Let activation of layer k be $z_k$, 
 All m layers in the network, $ q_{m+1} = \mathit{I} $
 
$$ q_k = (w q_{k+1})(z_k > 0) $$
 
$$ \nabla x = q_1 $$

```python
gw2_f = theano.function([x, w, b], gw2)
x_var = np.array([0.8, 0.2])
w_var = np.array([[-0.01,0.2],[0.9,0.03]])
b_var = np.zeros(2)
print "theano grad:", gw2_f(x_var, w_var, b_var)
print "hand-crafted grad:", np.vstack([x_var,x_var]).T * ((x_var.dot(w_var) + b_var) > 0)
```

### Gradients of long reccurent network

```python
x = T.vector("x")
b = T.vector("b")
w = T.matrix("w")
n = T.iscalar("n")

def recurent(s):
    return relu(T.dot(s, w) + b)

z, _ = theano.scan(recurent, outputs_info=[x], n_steps=n)

gw = T.grad(T.sum(z), w)
gw_f = theano.function([x, w, b, n], gw)
```

```python
x_var = np.array([0.8, 0.2])
b_var = np.zeros(2)
```

#### Identity weight matrix with diagonal value of 0.9

```python
w_var = np.array([[0.9,0.0],[0.0,0.9]])

steps = np.arange(1, 101)
grad_norms = [np.linalg.norm(gw_f(x_var, w_var, b_var, n)) for n in steps]
plt.plot(steps, grad_norms)
plt.show()
```

#### Identity weight matrix with diagonal value of 1.0

```python
w_var = np.array([[1.0,0.0],[0.0,1.0]])

steps = np.arange(1, 101)
grad_norms = [np.linalg.norm(gw_f(x_var, w_var, b_var, n)) for n in steps]
plt.plot(steps, grad_norms)
plt.show()
```

#### random weight matrix

```python
w_var = np.array([[0.03,-0.2],[0.3,-0.05]])

steps = np.arange(1, 101)
grad_norms = [np.linalg.norm(gw_f(x_var, w_var, b_var, n)) for n in steps]
plt.plot(steps, grad_norms)
plt.show()
```

## tanh as activation

```python
x = T.vector("x")
b = T.vector("b")
w = T.matrix("w")
n = T.iscalar("n")

def recurent(s):
    return T.tanh(T.dot(s, w) + b)

z, _ = theano.scan(recurent, outputs_info=[x], n_steps=n)

gw = T.grad(T.sum(z), w)
gw_f = theano.function([x, w, b, n], gw)
```

### Identity weight matrix of 0.9

```python
w_var = np.array([[0.9,0.0],[0.0,0.9]])

steps = np.arange(1, 101)
grad_norms = [np.linalg.norm(gw_f(x_var, w_var, b_var, n)) for n in steps]
plt.plot(steps, grad_norms)
plt.show()
```

### Identity weight matrix of 1.0

```python
w_var = np.array([[1.0,0.0],[0.0,1.0]])

steps = np.arange(1, 101)
grad_norms = [np.linalg.norm(gw_f(x_var, w_var, b_var, n)) for n in steps]
plt.plot(steps, grad_norms)
plt.show()
```

### Random weight matrix

```python
w_var = np.array([[0.03,-0.2],[0.3,-0.05]])

steps = np.arange(1, 101)
grad_norms = [np.linalg.norm(gw_f(x_var, w_var, b_var, n)) for n in steps]
plt.plot(steps, grad_norms)
plt.show()
```
