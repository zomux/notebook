### Theano convolution test

```python
import numpy as np
import theano
import theano.tensor as T
```

```python
_x = T.tensor3()
_w = T.tensor4()

x4 = _x.dimshuffle(("x", 0, 1, 2))

result = T.nnet.conv2d(x4, _w,  border_mode="valid")

f = theano.function([_x, _w], result)
```

```python
x = np.array([
    [[3, 2, 5, 6, 7, 8],
      [5, 4, 2, 10, 8, 1],
    [1, 1, 2, 1, 1, 2]],
        
    [[1, 1, 2, 1, 1, 2],
      [1, 1, 2, 1, 1, 2],
    [1, 1, 2, 1, 1, 2]]
    ], dtype="float32")

w = np.array([
[  
        [  [4, 5, 3],
           [1, 2, 3]
        ],
        [   [1, 1, 1],
            [1, 1, 2]
        ]
             
],
[  
        [  [1, 1, 3],
           [1, 1, 2]
        ],
        [   [1, 1, 1],
            [1, 1, 2]
        ]
             
]], dtype="float32")


f(x, w)
```
