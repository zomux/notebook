# Metropolis-Hastings

```python
from numpy import *
from scipy import linalg as LA
from matplotlib.pyplot import *
import matplotlib.cm as cm
```

```python
mu = array([0, 0]) # 平均
S  = array([[1, 0.5],[0.5, 1]]) # 分散
Sinv = LA.inv(S)
detS = LA.det(S) # determinant
```

```python
S
```

## Multivariant normal Distr.
![](http://upload.wikimedia.org/math/7/1/a/71a05f280112cf66ce2c8641f9b898ab.png)

```python
def f(x):
    return exp(-(x-mu).T.dot(Sinv).dot(x-mu)/2)/(2*pi*sqrt(detS))
```

### Draw grid

```python
X, Y = meshgrid(linspace(-3, 3, 100), linspace(-3, 3, 100))
Z = vectorize(lambda x,y: f([x,y]))(X, Y)

xlim(-3, 3)
ylim(-3, 3)
pcolor(X, Y, Z, alpha=0.3)
show()
```

## MH:
- The proposed distr. is q
-  Accept the sample with following prob
 $$ \mathrm{min}\left\{\frac{q(\mathbf{x}^{(t)}|\mathbf{x}^{(t+1)})\pi(\mathbf{x}^{(t+1)})}{q(\mathbf{x}^{(t+1)}|\mathbf{x}^{(t)})\pi(\mathbf{x}^{(t)})}, 1 \right\}$$

## Random Walk
$$\mathbf{x}^{(t+1)} = \mathbf{x}^{(t)} + \mathbf{z}$$
### Where z is normal distr. or t distr.
### In random walk $q(x^\prime|x) = q(x|x^\prime)$, so,
$$p_{next} = \mathrm{min}\left\{\frac{\pi(\mathbf{x}')}{\pi(\mathbf{x})}, 1 \right\}$$
### Here, z is $\mathcal{N}(\mathbf{0}, \sigma^2I)$

```python

# MH法
sigma=1
def next(x):
    while True:
        new_x = x + random.normal(0, sigma, 2)
        if random.uniform() <= min( f(new_x)/f(x), 1 ):
            return new_x

BURNIN = 10 # グラフの見やすさの為に小さな値にしています
N = 100

x = [2, -2]
```

```python

# バーンイン
burn_x = zeros(BURNIN); burn_y = zeros(BURNIN)
for i in range(BURNIN):
    burn_x[i] = x[0]
    burn_y[i] = x[1]
    x = next(x)

# サンプリング
sample_x = zeros(N); sample_y = zeros(N)
```

```python

# グラフの見た目を良くするために１点共有
sample_x[0] = burn_x[-1]
sample_y[0] = burn_y[-1]

for i in range(1, N):
    sample_x[i] = x[0]
    sample_y[i] = x[1]
    x = next(x)

xlim(-3, 3)
ylim(-3, 3)
pcolor(X, Y, Z, alpha=0.3)
plot(burn_x, burn_y, label="burn-in")
plot(sample_x, sample_y, label="sampling")
legend(loc=1)
show()
```
