auto clip features

```python
cliped_feats = []
for i in range(X_feature.shape[1]):
    feat = X_feature[:, i] 
    max_v = np.mean(feat) + np.std(feat) * 7
    min_v = np.mean(feat) - np.std(feat) * 7
    new_feat = max_v * (feat > max_v) + min_v * (feat < min_v) + feat * (feat <= max_v) * (feat >= min_v)
    cliped_feats.append(new_feat)
cliped_feats = np.vstack(cliped_feats).T
```

```python
plt_x = np.arange(cliped_feats.shape[0])
for i in range(cliped_feats.shape[1]):
    plt.subplot(cliped_feats.shape[1], 3, i+1)
    plt.plot(plt_x, cliped_feats[:, i])
    plt.ylabel("plt " + str(i))
plt.subplots_adjust(top=20, left=10, right=11)
plt.show()
```
