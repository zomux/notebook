# Normal Distribution

```python
from numpy import *
from pylab import *
```

```python
xs = linspace(-10, 10)
```

```python
mu = 3
sigma = 5
```

![](http://upload.wikimedia.org/math/1/f/b/1fbab0f6399a52a3cfec041b458e62a1.png)

```python
def fn(mu, sigma, x):
    return (1 / sqrt(2 * 3.1415 * (sigma **2))) * exp( - ((x - mu)**2) / (2 * sigma**2) )
```

```python
import random
ys = array([fn(mu, sigma, x) + (random.random()-0.5)*0.05 for x in xs])
```

```python
scatter(xs, ys, s=20)
```

# Mean Squared Error: $$ J(\mu, \sigma) = \sum_i (fn(\mu, \sigma, x_i) - y_i)^2$$

```python
def j(arr):
    mu, sigma = arr
    total = 0.0
    for i in range(len(xs)):
        x = xs[i]
        y = ys[i]
        total += (fn(mu, sigma, x) - y) ** 2
    return total
```

```python
mu_0, sigma_0 = 1, 1 
```

```python
from scipy.optimize import fmin_bfgs
mu_final, sigma_final = fmin_bfgs(j, array([mu_0, sigma_0]))
```

```python
mu_final, sigma_final
```

```python
scatter(xs, ys, s=20)
plot(xs, array([fn(mu_final, sigma_final, x) for x in xs]), "r-")
```
