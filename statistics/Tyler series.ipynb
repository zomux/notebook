### sin

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
import math

def f_sin(x):
    return x - x**3/math.factorial(3) + x**5/math.factorial(5) - x**7/math.factorial(7) \
            + x**9/math.factorial(9) - x**11/math.factorial(11)
```

```python
xspace = np.linspace(-6,6)
yspace = map(f_sin, xspace)
plt.plot(xspace, yspace, label="Tyler approx.")
plt.plot(xspace, map(math.sin, xspace), label="sin(x)")
plt.legend()
plt.show()
```

### e^x

```python
import math

def f_e(x, k):
    sum = 1.0
    for n in range(1, k + 1):
        sum += x**n / math.factorial(n)
    return sum

xspace = np.linspace(0,7)
for k in range(1, 7):
    plt.plot(xspace, [f_e(x, k) for x in xspace], label="%d-order Tyler approx." % k)
plt.plot(xspace, map(math.exp, xspace), label="e^x")
plt.legend()
plt.show()
```
