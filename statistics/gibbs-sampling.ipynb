## 2D Multivariate normal distribution

```python
rho = 0.5
mu = array([0, 0]) # 平均
S  = array([[1, rho],[rho, 1]]) # 分散
Sinv = LA.inv(S)
detS = LA.det(S) # determinator
print "S=", S
print "detS=", detS
```

## The formula is: ![](http://upload.wikimedia.org/math/7/1/a/71a05f280112cf66ce2c8641f9b898ab.png)

* So, $|\sum|$ is Determinant here, which is calculated by $1 \times 1 - 0.5 \times 0.5$
* In 2D case, $k = 2$

## In this case: $$f(x) = \frac{1}{2\pi\sqrt{detS}} \exp((-\frac{1}{2}(x - mu)^T S^{-1} (x - mu)))$$

```python
def f(x):
    return exp(-(x-mu).T.dot(Sinv).dot(x-mu)/2)/(2*pi*sqrt(detS))

X, Y = meshgrid(linspace(-3, 3, 100), linspace(-3, 3, 100))
Z = vectorize(lambda x,y: f([x,y]))(X, Y)

xlim(-3, 3)
ylim(-3, 3)
pcolor(X, Y, Z, alpha=0.3)
show()
```

## Do Gibbs sampling

* Compute new x based on y
* Compute new y based on new x

* Random normal distr. assume x = 1

```python
tmpdistr = [random.normal(rho*1, 1-rho**2) for x in range(1000)]
plot(tmpdistr, "k.")
show()
```

```python
# Gibbs sampling
def next(x):
    new_x = random.normal(rho*x[1], 1-rho**2)
    new_y = random.normal(rho*new_x, 1-rho**2)
    return [new_x, new_y]
```

## Burn In

```python
BURNIN = 10 # グラフの見やすさの為に小さな値にしています
N = 100

x = [2, -2]

# バーンイン
burn_x = zeros(BURNIN); burn_y = zeros(BURNIN)
for i in range(BURNIN):
    burn_x[i] = x[0]
    burn_y[i] = x[1]
    x = next(x)
```

```python

# サンプリング
sample_x = zeros(N); sample_y = zeros(N)

# グラフの見た目を良くするために１点共有
sample_x[0] = burn_x[-1]
sample_y[0] = burn_y[-1]

for i in range(1, N):
    sample_x[i] = x[0]
    sample_y[i] = x[1]
    x = next(x)

xlim(-3, 3)
ylim(-3, 3)
pcolor(X, Y, Z, alpha=0.3)
plot(burn_x, burn_y, label="burn-in")
plot(sample_x, sample_y, label="sampling")
legend(loc=1)
show()
```

```python
xs = np.linspace(1,10,100)
ys = gamma(xs)
plot(xs, ys, "k-")
p = polyfit(xs, ys, 2)
ps = xs**2*p[0] + xs*p[1] + p[2]
plot(xs, ps, "r-")
show()
```

```python
# ディリクレ分布
a = [3,2,5]
z = sum([gamma(x) for x in a])/gamma(sum(a))
def p(x):
    if x[0] + x[1] > 1:
        return -0.1
    return x[0]**a[0] * x[1]**a[1] * (1-x[0]-x[1])**a[2] / z
```

```python
X, Y = meshgrid(linspace(0, 1, 100), linspace(0, 1, 100))
Z = vectorize(lambda x,y: p([x,y]))(X, Y)

xlim(0, 1)
ylim(0, 1)
pcolor(X, Y, Z, alpha=0.3)
title("Dirichlet distribution a=%s" % a)
show()
```

## Gibbs Sampling

```python

# Gibbs sampling
def next(x):
    new_x = random.beta(a[0]+1, a[2]+1)*(1-x[1])
    new_y = random.beta(a[1]+1, a[2]+1)*(1-new_x)
    return [new_x, new_y]

BURNIN = 10 # グラフの見やすさの為に小さな値にしています
N = 100

x = [0, 0]

# バーンイン
burn_x = zeros(BURNIN); burn_y = zeros(BURNIN)
for i in range(BURNIN):
    burn_x[i] = x[0]
    burn_y[i] = x[1]
    x = next(x)

# サンプリング
sample_x = zeros(N); sample_y = zeros(N)

# グラフの見た目を良くするために１点共有
sample_x[0] = burn_x[-1]
sample_y[0] = burn_y[-1]

for i in range(1, N):
    sample_x[i] = x[0]
    sample_y[i] = x[1]
    x = next(x)

xlim(0, 1)
ylim(0, 1)
pcolor(X, Y, Z, alpha=0.3)
plot(burn_x, burn_y, label="burn-in")
plot(sample_x, sample_y, label="sampling")
title("Dirichlet distribution a=%s" % a)
legend(loc=1)
show()
```
