```python
from numpy import *
from scipy import stats
from scipy import linalg as LA
from matplotlib.pyplot import *
import matplotlib.cm as cm
from StringIO import StringIO
```

## Data from $y_i = ax_i + b + \epsilon_i $
* Noise $\epsilon \sim \mathcal{N}(0, \sigma^2)$
* a, x
$$\mathbf{a}=(a,b)^T,\quad
        \mathbf{X}= \begin{pmatrix}
        x_1 & x_2 & \cdots & x_n \\
        1   &   1 & \cdots &   1
        \end{pmatrix}^T$$
* y
$$\mathbf{y} \sim \mathcal{N}(\mathbf{X}\mathbf{a},\sigma^2I)$$
* So $\mathbf{a},\sigma$ is the parameters of this distr.
* Try to estimate the post distr. of this two parameters when D is given as follows 

## The likehood when D is given
$$\begin{aligned}
        L(\mathbf{a},\sigma^2|D)&=\mathcal{N}(\mathbf{y}|\mathbf{X}\mathbf{a},\sigma^2I)\\
        & \propto (\sigma^2)^{-n/2}\exp\left\{-\frac{1}{2\sigma^2}||\mathbf{y}-\mathbf{X}\mathbf{a}||^2\right\}
        \end{aligned}$$

## Posterior Distr. of parameters

$$\pi(\mathbf{a},\sigma^2|D)\propto L(\mathbf{a},\sigma^2|D)\mathcal{N}(\mathbf{a}|\mathbf{a}_0,\sigma_0^2I)IG(\sigma^2|\alpha_0/2,\beta_0/2)$$

* Which comes from beyesian theorem
$$$$
$$\pi(\mathbf{a},\sigma^2|D) \propto \mathcal{N}(D|\mathbf{a},\sigma^2) p(\mathbf{a},\sigma^2)$$
* Where,
$$p(\mathbf{a},\sigma^2) = \mathcal{N}(\mathbf{a}|\mathbf{a}_0,\sigma_0^2I)IG(\sigma^2|\alpha_0/2,\beta_0/2)$$
$$\mathcal{N}(D|\mathbf{a},\sigma^2) = L(\mathbf{a},\sigma^2|D) $$ 
* So for Gibbs sampling, when we fix $\sigma$
\begin{aligned}
        \pi(\mathbf{a}|\sigma^2,D)&\propto  L(\mathbf{a},\sigma^2|D)\mathcal{N}(\mathbf{a}|\mathbf{a}_0,\sigma_0^2I) \\
        &\propto \exp\left\{-\frac{1}{2\sigma^2}||\mathbf{y}-\mathbf{X}\mathbf{a}||^2-\frac{1}{2\sigma_0^2}||\mathbf{a}-\mathbf{a}_0||^2\right\} \\
        \end{aligned}
* Maximize this post prob, then
$$\begin{aligned}
        & \pi(\mathbf{a}|\sigma^2,D) = \mathcal{N}(\mathbf{a}', \mathbf{\Sigma}) \\
        & \mathbf{\Sigma}^{-1} = \frac{1}{\sigma^2}\mathbf{X}^T\mathbf{X}+\frac{1}{\sigma_0^2}I\\
        & \mathbf{a}' = \mathbf{\Sigma}\left(\frac{1}{\sigma^2}\mathbf{X}^T\mathbf{y}+\frac{1}{\sigma_0^2}\mathbf{a}_0\right)
        \end{aligned}$$
* When fix $a$
\begin{aligned}
        \pi(\sigma^2|\mathbf{a},D) &\propto L(\mathbf{a},\sigma^2|D)IG(\sigma^2|\alpha_0/2,\beta_0/2) \\
        &\propto (\sigma^2)^{-n/2-\alpha_0/2-1}\exp\left\{-\frac{1}{2\sigma^2}||\mathbf{y}-\mathbf{X}\mathbf{a}||^2-\frac{1}{2\sigma^2}\beta_0\right\}
        \end{aligned}
* then
$$\begin{aligned}
        &\pi(\sigma^2|\mathbf{a},D) = IG(\alpha/2,\beta/2) \\
        &\alpha = \alpha_0 + n\\
        &\beta  = \beta_0 + ||\mathbf{y}-\mathbf{X}\mathbf{a}||^2
        \end{aligned}$$

```python
# aの事前分布のパラメータ
a0  = array([0,0])
s02 = 1.0e2 ** 2

# σ^2の事前分布のパラメータ
alpha0 = 1.0e-2
beta0  = 1.0e2
```

```python
x = data[:,0]
y = data[:,1]
n = len(x)
```

```python
# 繰り返し使う行列・ベクトルを作っておく
X = array([x, ones(n)]).T
XtX = X.T.dot(X)
Xty = X.T.dot(y)
```

```python
# π(a|σ^2,D)
def p_a(s2):
    Sigma  = LA.inv(XtX/s2 + 1/s02)
    a_dash = Sigma.dot(Xty/s2 + a0/s02)
    return random.multivariate_normal(a_dash, Sigma)

# π(σ^2|a,D)
def p_s2(a):
    alpha = alpha0 + n
    beta  = beta0  + LA.norm(y-X.dot(a))**2
    # X ～ IG(a,b)  <=>  1/X ～ Ga(a,1/b)
    return 1/random.gamma(alpha/2, 2/beta)

# ギブスサンプラー
def sampler(BURNIN):
    a  = array([0, 0])  # 係数の初期値
    s2 = 1.0            # 分散の初期値
    for i in range(BURNIN):
        a  = p_a(s2)
        s2 = p_s2(a)
    while True:
        a  = p_a(s2)
        s2 = p_s2(a)
        yield a[0], a[1], s2

sim = sampler(1000)

NSAMPLE = 10000
a_samples = zeros(NSAMPLE)
b_samples = zeros(NSAMPLE)
s2_samples = zeros(NSAMPLE)
for i in range(NSAMPLE):
    a_samples[i], b_samples[i], s2_samples[i] = sim.next()

# 事後分布の図示
def posterior(name, samples):
    NBINS = 100
    title(u"posterior distribution of %s (N=%d, mean=%.2f, std=%.2f)"
            % (name, NSAMPLE, average(samples), std(samples)))
    hist(samples, bins=NBINS, normed=True)
    show()

posterior("a", a_samples)
posterior("b", b_samples)
posterior(u"σ^2", s2_samples)

#MAP 推定値
a_MAP  = average(a_samples)
b_MAP  = average(b_samples)
s2_MAP = average(s2_samples)

# 回帰直線の表示
xmin = 0;  xmax = 1
ymin = -1; ymax = 5
xlim(xmin, xmax)
ylim(ymin, ymax)
t = linspace(0, 1, 100)
regress_line = a_MAP*t + b_MAP
xlim(0, 1)
scatter(x, y)
plot(t, regress_line)
fill_between(t, regress_line+sqrt(s2_MAP), regress_line-sqrt(s2_MAP),
        alpha=0.2)
title(u"regression line")
show()

# 予測分布
def predictive(sim, x):
    a,b,s2 = sim.next()
    return random.normal(a*x+b, sqrt(s2))
```

```python

xlim(xmin, xmax)
ylim(ymin, ymax)
xs = linspace(xmin, xmax, 100)
means = []
ys_high = []
ys_low  = []
for t in xs:
    M = 100
    mean = a_MAP*t + b_MAP
    s = sqrt(average([(mean - predictive(sim, t))**2 for i in range(M)]))
    means.append(mean)
    ys_high.append(mean + s)
    ys_low.append(mean - s)

scatter(x, y)
plot(xs, means)
plot(xs, ys_high)
plot(xs, ys_low)
show()
```
